# Data Collection

```python
import pandas as pd

df = pd.read_csv("hf://datasets/GerindT/mini_amazon_sentimental/encoded_test_dataset.csv")
```

```python
pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
df["sentiment"].value_counts(normalize= True )
```

```python
natural_size = int(0.076860*df.shape[0])
natural_size
```

```python
df_Neutral = df[df['sentiment'] == 'Neutral']
df_Negative = df[df['sentiment'] == 'Negative'].head(natural_size)
df_Perfect = df[df['sentiment'] == 'Perfect'].head(natural_size)
df_Satisfied  = df[df['sentiment'] == 'Satisfied'].head(natural_size)
dataset = pd.concat([df_Neutral, df_Negative, df_Perfect, df_Satisfied])
dataset
```

# EDA

```python
# check that the data is balanced
dataset["sentiment"].value_counts(normalize= True )
```

```python
dataset.drop(["title","content"],axis = 1,inplace=True)
```

```python
dataset['text'] = dataset['text'].str.lower()
dataset.tail()
```

```python
dataset['char_count'] = dataset['text'].str.len()
dataset['word_count'] = dataset['text'].str.split().str.len()
dataset['avg_word_length'] = dataset['char_count'] / dataset['word_count']
```

```python
dataset.head()
```

```python
dataset['star'] = dataset['label'].apply(lambda x : x.split()[1][1])
dataset.head()
```

```python
dataset.drop(['label'],axis = 1,inplace=True) 
dataset.head()
```

```python
import random

# Initialize an empty dictionary
tweet_dict = {}

# Get the number of rows in the dataset
num_rows = len(dataset)

for i in range(10):
    # Generate a random index within the range of dataset rows
    ind = random.randint(0, num_rows - 1)
    # Store the tweet in the dictionary with the index as the key
    tweet_dict[ind] = dataset.iloc[ind]  # Use .iloc to access by index

# Print or return the dictionary
print(tweet_dict)
```

```python
dataset['text'] = dataset['text'].str.replace('[^a-zA-Z]',' ',regex=True)
dataset.head()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
dataset['text'] = dataset['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
dataset.head()
```

```python
import spacy

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Define a function to lemmatize text
def lemmatize_text(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc]

# Apply the lemmatization function to your dataset
dataset['text'] = dataset['text'].apply(lemmatize_text)
```
